```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np


# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
combined_df = pd.merge(ride_data, city_data, how="left")

# Display the data table for preview
combined_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
urban_df = combined_df.loc[combined_df["type"] == "Urban", :]
suburban_df = combined_df.loc[combined_df["type"] == "Suburban", :]
rural_df = combined_df.loc[combined_df["type"] == "Rural", :]
urban_df
```

```python
# Obtain the x and y coordinates for each of the three city types

#x and y coordinates for Urban
x_urban = urban_df["city"].value_counts()
y_urban = urban_df.groupby("city")["fare"].mean()
s_urban = urban_df.groupby("city")["driver_count"].sum()


#x and y coordinates for Suburban
x_suburban = suburban_df["city"].value_counts()
y_suburban = suburban_df.groupby("city")["fare"].mean()
s_suburban = suburban_df.groupby("city")["driver_count"].sum()

#x and y coordinates for Rural
x_rural = rural_df["city"].value_counts()
y_rural = rural_df.groupby("city")["fare"].mean()
s_rural = rural_df.groupby("city")["driver_count"].sum()

# Build the scatter plots for each city types
# Incorporate the other graph properties
plt.ylim(18, 45)
plt.xlim(0, 42)
plt.grid()
plt.title("Pyber Ride Sharing Data (2016)")
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)")
plt.text(.5, .5, "Note:Circle size correlates with driver count per city.")
Urban = plt.scatter(x_urban, y_urban, s_urban, color="lightcoral", edgecolors="black", alpha=0.7, linewidth=.7)
Suburban = plt.scatter(x_suburban, y_suburban, s_suburban, color="lightskyblue", edgecolors="black", alpha=0.7, linewidth=.7)
Rural = plt.scatter(x_rural, y_rural, s_rural, color="gold", edgecolors="black", alpha=0.7, linewidth=.7)

# Create a legend
label = ["Urban", "Suburban", "Rural"]
plt.legend(labels = label, handles = [Urban, Suburban, Rural], title = "City Types", loc="best")

# Incorporate a text label regarding circle size

# Save Figure
plt.savefig("../Images/PyberBubble.png")
```

```python
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
fare_total = combined_df["fare"].sum()
fare_urban = (urban_df["fare"].sum()/fare_total) * 100
fare_suburban = (suburban_df["fare"].sum()/fare_total) * 100
fare_rural = (rural_df["fare"].sum()/fare_total) * 100
# Build Pie Chart
types = ["Urban", "Suburban", "Rural"]
fare_types = [fare_urban, fare_rural, fare_suburban]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(fare_types, explode=explode, labels=types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=290, radius=1.5)
plt.title("% of Total Fares by City Type", pad=40)

# Save Figure
plt.savefig("../Images/FarePie.png")
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python
# Calculate Ride Percents
ride_total = len(combined_df["ride_id"])
ride_urban = len(urban_df["ride_id"])/ride_total * 100
ride_suburban = len(suburban_df["ride_id"])/ride_total * 100
ride_rural = len(rural_df["ride_id"])/ride_total * 100
# Build Pie Chart
types = ["Urban", "Suburban", "Rural"]
fare_types = [ride_urban, ride_rural, ride_suburban]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(fare_types, explode=explode, labels=types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=270, radius=1.5)
plt.title("% of Total Rides by City Type", pad=40)
# Save Figure
plt.savefig("../Images/RidePie.png")
```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python
# Calculate Driver Percents
driver_total = combined_df["driver_count"].sum()
driver_urban = s_urban.sum()/driver_total * 100
driver_suburban = s_suburban.sum()/driver_total * 100
driver_rural = s_rural.sum()/driver_total * 100
# Build Pie Charts
types = ["Urban", "Suburban", "Rural"]
fare_types = [driver_urban, driver_rural, driver_suburban]
colors = ["lightcoral", "gold", "lightskyblue"]
explode = (0.1, 0, 0)
plt.pie(fare_types, explode=explode, labels=types, colors=colors, autopct="%1.1f%%", shadow=True, startangle=290, radius=1.5)
plt.title("% of Total Drivers by City Type", pad=60)
# Save Figure
plt.savefig("../Images/DriverPie.png")
```

```python
# Show Figure
plt.show()
```
